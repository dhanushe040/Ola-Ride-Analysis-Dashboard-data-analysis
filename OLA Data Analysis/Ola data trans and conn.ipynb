```python
df['Reason for Cancelling by Customer'] = df['Reason for Cancelling by Customer'].fillna('Not Cancelled by Customer')
df['Reason for Cancelling by Driver'] = df['Reason for Cancelling by Driver'].fillna('Not Cancelled by Driver')
df['Incomplete Rides Reason'] = df['Incomplete Rides Reason'].fillna('Completed Ride')
```

```python
df.isnull().sum()
```

```python
for col in ['Booking Value', 'Ride Distance', 'Driver Ratings', 'Customer Rating', 'Avg VTAT', 'Avg CTAT']:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df['Payment Method'] = df['Payment Method'].fillna('Not Applicable')
```
